```python
# import pandas
import pandas as pd

# loading data into DataFrame
pokemon = pd.read_csv('Pokemon.csv')

# printing out some of the dataset
print(pokemon.head(5))
```

```python
# import seaborn
import seaborn as sns
# import seaborn
import seaborn as sns

# simple scatter plot
sns.scatterplot(x='Speed', y='Sp. Atk', data=pokemon)
```

```python
column = "Speed"

sns.histplot(data=pokemon, x=column)
```

```python
sns.lineplot(data=pokemon, x="Total", y="Speed")
# pokemon["Total"].to_numpy()
```

```python
#Create a bar plot of the number of legendary and non-legendary Pokemon using the seaborn countplot() function.

import matplotlib.pyplot as plt

# read a tips.csv file from seaborn library


# count plot on single categorical variable
sns.countplot(x ='Legendary', data = pokemon)

plt.show()
```

```python
data = pokemon[["HP", "Sp. Atk", "Sp. Def", "Speed"]]
sns.heatmap(data.corr())
```
